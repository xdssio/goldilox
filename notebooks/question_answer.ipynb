```python
from transformers import pipeline as QustionAswering
import vaex
import pyarrow as pa
from goldilox import Pipeline

qa = QustionAswering('question-answering')


@vaex.register_function()
def answer(ar):
    if isinstance(ar, str):
        ar = [ar]    
    if hasattr(ar, 'tolist'):
        ar = ar.tolist()
    ret = pa.array(qa(question, story)['answer'] for question in ar)
    return ret


df = vaex.from_arrays(question=['What do you mean by by "It takes two to tango"?'])
df.add_function("answer", answer)
df['question'] = df['question'].fillna('')
df['results'] = df.question.answer()
model = Pipeline.from_vaex(df)
model.inference({'question': 'Familiarity breeds contempt'})
```
